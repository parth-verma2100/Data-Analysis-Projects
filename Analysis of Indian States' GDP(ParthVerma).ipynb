```python
import pandas
from bs4 import BeautifulSoup
import time
from selenium import webdriver
import unicodedata
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_Indian_states_and_union_territories_by_GDP")
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
soup
```

```python
table = soup.find_all('table', class_="wikitable")[2]
```

```python
table_headers = table.find_all('th', class_='headerSort')
table_headers
```

```python
years_list = []
for i in table_headers:
    years_list.append(i.text)
```

```python
data_replaced = [unicodedata.normalize('NFKD', item).encode('ascii', 'replace').replace('?', '-') for item in years_list]
```

```python
gdp_years = [item.replace("\n", '') for item in data_replaced]
```

```python
gdp_years
```

```python
table_body = table.find('tbody')
table_body
```

```python
states_yearly_records = table_body.find_all('tr')
```

```python
states_gdp_records = []
for i in states_yearly_records:
    data = []
    state_name = unicodedata.normalize('NFKD', i.find('a').text).encode('ascii', 'ignore').replace(" *", "")
    data.append(state_name)
    for j in i.find_all('span'):
        data.append(unicodedata.normalize('NFKD', j.text).encode('ascii', 'ignore'))
    states_gdp_records.append(data)
gdp_records = list(map(lambda row: list(filter(lambda x: x != '', row)), states_gdp_records))
```

```python
gdp_records
```

```python
df=pandas.DataFrame(gdp_records, columns=gdp_years)
```

```python
df
```

```python
df.to_csv('table_data.csv', index=False)
```
