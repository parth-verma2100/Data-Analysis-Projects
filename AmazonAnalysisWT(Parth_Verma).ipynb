```python
import pandas as pd
```

```python
df = pd.read_csv("Reviews.csv")
```

```python
df
```

```python
# Q1: What is the distribution of review scores?
# Ans: According to below analysis, we can see that mostly users are liking the product as we have majority of
#      5 out of 5 given by 3.6L customers which is a positive feedback from customers point of view.

df['Score'].value_counts()
```

```python
# Q2: How does the helpfulness numerator compare to the helpfulness denominator across reviews?
# Ans: After analysing below output we can conclude that the Helpfullness Numerator and Denominator are varying with 
#      rating as we can see for Top rating i.e. 5 we are having less Helpfullness Numerator than Denominator 
#      similarly as we go down the chart to below rating i.e. 1 we are having more Helpfullness Numerator than
#       Denominator so as rating gets down we can conclude we are having less Helpfullness Numerator down the line
#      comparing to the Denominator so they are inversly proportional

data = df.groupby('Score')
num_count = data['HelpfulnessNumerator'].sum()
den_count = data['HelpfulnessDenominator'].sum()

pd.DataFrame({'HelpfulnessNumerators': num_count,'HelpfulnessDenominators': den_count })
```

```python
# Q3: What are the average review scores for different products?
# Ans: As per the below analysis we can see that for each product mostly users are giving a positive feedback
#      5 out of 5 rating and very less users are giving below 5 so overall result is positive from the output.

df.groupby('ProductId')['Score'].value_counts()
```

```python
# Q4: Is there a correlation between the length of the review text and the review score?
# Ans: As per the analysis, we can see that for EXCELLENT i.e. 5 rating users write 100-800 length of text whereas 
#        as move to below rating to POOR user providing review in 100-2000 length of text to justify issues.

filtered_text = df['Text'].apply(lambda x: len(x))
mdf = pd.DataFrame({'Score': df['Score'], 'TextLen': filtered_text})
mdf.sort_values('Score', ascending=False)
```

```python
# Q5: Who are the top 10 most active reviewers, and how many reviews have they left?
# Ans: If we check the below analysis we get the top 10 users with max reviews and their counts just after the 
# text column.

df.groupby('UserId')['Text'].count().sort_values(ascending=False).head(10)
```

```python
# Q6: How has the average length of reviews changed over time?
# Ans: As we can see from below chart that as the time increase from prev years to current
#      there has been a vast increment in the length of the reviews hence we can say that
#      as time increase the lenght of text reviews also increase

date_time = pd.to_datetime(df['Time'], unit='s')
len_t = df['Text'].apply(len)

pd.DataFrame({'Time': date_time, 'Length': len_t}).sort_values('Length')
```

```python
# Q7: Do reviews with higher scores receive more helpfulness votes?
# Ans: From below analysis we can see that the helpfull ness votes are highest for thw reviews having score 5
#      through which we can conclude that reviews with higher scores are recieving more helpfullness votes.

data = df[['HelpfulnessNumerator', 'Score']].groupby('Score').sum().reset_index()
data.columns = ['Score', 'H.Votes']
data
```

```python
# Q8: Is there a relationship between the number of reviews a product has and its average score?
# Ans: As we can see from below analysis, when were are having the average score around 4 then we have 
#      maximum of views other than that for products having average score as 5 mostly we have only 1 view and
#      similarly for average score lesser than 4 we have nearby 1 or 2 views

group_data = df.groupby('ProductId')
score = group_data['Score'].mean()
views = group_data['Text'].count()
result = pd.DataFrame({'AverageScore': score, 'Views': views}).reset_index()
result
```

```python
# Q9: How does the length of negative reviews compare to the length of positive reviews?
# Ans: From below analysis, we can see than the negative reviews length lies in range from
#      100-2000 whereas for positive reviews length lies in range from 100-800 so we can 
#      say that positive reviews have lesser length than negative reviews.

filtered_text = df['Text'].apply(lambda x: len(x))
mdf = pd.DataFrame({'Score': df['Score'], 'TextLen': filtered_text})
mdf.sort_values('Score', ascending=False)
```

```python
# Q10: What are the average review scores and helpfulness ratios for the top 10 products?
# Ans: As per below analysis we can see that generally for the product having high scores are having more helpfull
#      reviews as compare to products having lesser scores which means customers are not clearly providing
#      justification to the quality of product due to which company is lagging to know the actual reason behing
#      product poor quality rating.

data = df.groupby('ProductId')
avg_scores = data['Score'].mean()
df['HelpfulnessRatio'] = df['HelpfulnessNumerator'] / df['HelpfulnessDenominator']
average_help = data['HelpfulnessRatio'].mean()
new_df = pd.DataFrame({'Average Scores': avg_scores, 'Average Helpfullness': average_help})
new_df.sort_values('Average Scores', ascending=False)
```
